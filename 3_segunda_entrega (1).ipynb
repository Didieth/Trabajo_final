```python
import pandas as pd
import matplotlib.pyplot as plt
datos = pd.read_excel("/content/drive/MyDrive/SEGUNDA_ENTREGA_PROG/HURTO_VEHICULOS.xlsx")
datos['FECHA_HECHO'] = pd.to_datetime(datos['FECHA_HECHO'])
datos['AÑO'] = datos['FECHA_HECHO'].dt.year
datos['MES'] = datos['FECHA_HECHO'].dt.month
pivot_table = datos.pivot_table(index='MES', columns='AÑO', values='CANTIDAD', aggfunc='sum')


fig, ax = plt.subplots(figsize=(20, 10))
pivot_table.plot(ax=ax, marker='o')

plt.xlabel('Mes')
plt.ylabel('Cantidad de incidentes')
plt.title('Cantidad de incidentes por mes y año')

plt.legend(title='Año')

for año in pivot_table.columns:
    for mes, cantidad in enumerate(pivot_table[año], start=1):
        offset = 0
        if cantidad > 0:
            offset = 10
        ax.annotate(str(int(cantidad)), (mes, cantidad), textcoords="offset points", xytext=(0, offset), ha='center')

meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
plt.xticks(range(1, 13), meses)

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

datos = pd.read_excel("/content/drive/MyDrive/SEGUNDA_ENTREGA_PROG/HURTO_VEHICULOS.xlsx")

datos = datos.dropna(subset=['BARRIOS_HECHO'])

cantidad_por_barrio = datos.groupby('BARRIOS_HECHO')['CANTIDAD'].sum()

bottom_20_barrios = cantidad_por_barrio.nsmallest(20).index

datos_bottom_20_barrios = datos[datos['BARRIOS_HECHO'].isin(bottom_20_barrios)]

cantidad_por_barrio = datos_bottom_20_barrios.groupby('BARRIOS_HECHO')['CANTIDAD'].sum()

plt.figure(figsize=(10, 8))
patches, texts, autotexts = plt.pie(cantidad_por_barrio, labels=cantidad_por_barrio.index, startangle=140,
                                    autopct=lambda p: '{:.0f}'.format(p * sum(cantidad_por_barrio) / 100))
plt.title('Cantidad de hechos por los 20 barrios con menos hechos')

for autotext in autotexts:
    autotext.set_fontsize(20)

plt.axis('equal')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

datos = pd.read_excel("/content/drive/MyDrive/SEGUNDA_ENTREGA_PROG/HURTO_VEHICULOS.xlsx")

datos['HORA_HECHO'] = pd.to_datetime(datos['HORA_HECHO'], format='%H:%M:%S')

datos['HORA'] = datos['HORA_HECHO'].dt.hour

frecuencia_por_hora = datos['HORA'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.bar(frecuencia_por_hora.index, frecuencia_por_hora.values, color='green')

for i, valor in enumerate(frecuencia_por_hora.values):
    plt.text(i, valor + 10, str(valor), ha='center', va='bottom', fontsize=9)

plt.title('Frecuencia de hurtos por hora del día')
plt.xlabel('Hora del día')
plt.ylabel('Frecuencia de hurtos')
plt.xticks(range(24))
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

datos = pd.read_excel("/content/drive/MyDrive/SEGUNDA_ENTREGA_PROG/HURTO_VEHICULOS.xlsx")

datos_filtrados = datos[['MODELO_VEHICULO', 'CANTIDAD']]

datos_agrupados = datos_filtrados.groupby('MODELO_VEHICULO')['CANTIDAD'].sum()

top_20_modelos = datos_agrupados.nlargest(20)

plt.figure(figsize=(12, 6))
bars = top_20_modelos.plot(kind='bar', color='black')

for index, value in enumerate(top_20_modelos):
    plt.annotate(str(value), xy=(index, value), ha='center', va='bottom')


plt.title('Top 20 modelos de vehículo con mayor cantidad de hurtos')
plt.xlabel('Modelo de vehículo')
plt.ylabel('Cantidad de hurtos')
plt.xticks(rotation=45, ha='right')

plt.show()
```
